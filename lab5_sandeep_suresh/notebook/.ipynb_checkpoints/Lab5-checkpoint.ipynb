```python
import pandas as pd
import numpy as np
from sklearn import tree
import graphviz
import statistics
from types import *
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
from scipy.stats import variation
import operator
import seaborn as sb
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Ridge
from sklearn.model_selection import KFold
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestRegressor
```

## Reading in Acute Inflamations Data and Transforming Dataframe

```python
df = pd.read_csv('../data/diagnosis.data', encoding='utf-16', sep='\s+', names=['Temp', 'Nausea', 'LumbarPain', 'Urine Pushing', 'Pains', 'Burning','Inflamation', 'Nephritis'])
df.head()
# Go through and change the temperate to using decimals instead of commas
for index, row in df.iterrows() :
    df.loc[index, 'Temp'] = row['Temp'].replace(",", ".")
    
df['Nausea'] = df['Nausea'].map({'yes' : 1, 'no' : 0})
df['LumbarPain'] = df['LumbarPain'].map({'yes' : 1, 'no' : 0})
df['Urine Pushing'] = df['Urine Pushing'].map({'yes' : 1, 'no' : 0})
df['Pains'] = df['Pains'].map({'yes' : 1, 'no' : 0})
df['Burning'] = df['Burning'].map({'yes' : 1, 'no' : 0})
df['Inflamation'] = df['Inflamation'].map({'yes' : 1, 'no' : 0})
df['Nephritis'] = df['Nephritis'].map({'yes' : 1, 'no' : 0})
df.head()
```

## Creating Decision Tree

```python
dTree = tree.DecisionTreeClassifier()
y = []
for index, row in df.iterrows() :
    y.append([row['Inflamation'], row['Nephritis']])
df.drop(['Inflamation', 'Nephritis'], axis=1)
dTree.fit(df, y)
```

```python
dot_data = tree.export_graphviz(dTree, out_file=None)
graph = graphviz.Source(dot_data)
```

## Reading in Crime Data

```python
dfCrime = pd.read_csv('../data/communities.data', header=None, prefix='X')
dropped = ['X0','X1','X2','X3','X4']
dfCrime.drop(dfCrime[dropped], axis=1, inplace=True)
dfCrime.columns = range(dfCrime.shape[1])
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
crimeCols = list(dfCrime)

# Changing the columns with strings to float
for col in crimeCols :

    # Check whether col contains strings
    # Find mean on whether it does or not
    if(isinstance(dfCrime[col][0], str)) :
        dfCrime[col] = pd.to_numeric(dfCrime[col], errors='coerce', downcast='float')
        
# Change nan to mean of each column
dfCrime = dfCrime.fillna(dfCrime.mean())
            
dfCrime.head()
```

```python
# Create and plot correlation matrix 
correlations = dfCrime.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
plt.show()
```

## Coefficient of Variation

```python
cvDict = {}
crimeCols = list(dfCrime)
for col in crimeCols :
    mean = statistics.mean(dfCrime[col])
    sd = statistics.stdev(dfCrime[col])
    cv = sd/mean
    cvDict[col] = cv
    
# Get the top 11 attributes
top11 = dict(sorted(cvDict.items(), key=operator.itemgetter(1), reverse=True)[:11])
print(top11)
```

## Making Scatter Plots

```python
for key,val in top11.items() :
    dfCrime.plot.scatter(x=key, y=122)
```

## Making Box Plots

```python
top11Keys = list(top11.keys())
boxplot = dfCrime.boxplot(column=top11Keys)
```

## Lasso Model 

```python
# Creating test and training set
trainingDf = dfCrime[:1495]
testDf = dfCrime[1495:]

X = trainingDf.iloc[:, range(122)]
y = np.array(trainingDf[122].values)

# Lasso Model picked with Cross validatoin and list of alphas
lassoMod = LassoCV(alphas=[10,1,0.5, 0.3, 0.1, 0.005, 0.001, 0.0005], cv=5) # 10 different alphas with 5 fold cross validation
lassoMod.fit(X,y)
coef = lassoMod.coef_
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")

# Outline which coefficients got picked
print("Coeffiecients Picked: ", end=" ")
for index in range(len(coef)) :
    if(coef[index] != 0) :
        print(index, end=" ")
        
print() # New Line
        
y_test = testDf.pop(122)
X_test = testDf
error_sum = 0
for index, row in X_test.iterrows() :
    y_pred = lassoMod.predict(row.values.reshape(1,-1))
    error = abs(y_pred - y_test[index])
    error_sum += error
print("Test Error: " + str((error_sum/len(X_test.index))[0]))
```

## Random Forests

```python
# Creating dataframe to store info
forestDfInfo = pd.DataFrame(columns=['m', 'B', 'Test-Error', 'OOB Error'])
# Creating test and training set
trainingDf = dfCrime[:1495]
testDf = dfCrime[1495:]

X = trainingDf.iloc[:, range(122)]
y = np.array(trainingDf[122].values)

y_test = testDf.pop(122)
X_test = testDf

# First m = sqrt(123) and m=123
for i in range(1,300) :
    forest = RandomForestRegressor(max_features="sqrt", n_estimators=i, warm_start=True, oob_score=True)
    forest.fit(X,y)

    # Calculate test error    
    error_sum = 0
    for index, row in X_test.iterrows() :
        y_pred = forest.predict(row.values.reshape(1,-1))
        error = abs(y_pred - y_test[index])
        error_sum += error
        
    size = len(X_test.index)
    average_error = error_sum/size
    forestDfInfo = forestDfInfo.append({'m' : 'sqrt', 'B' : i, 'Test-Error' : average_error[0], 'OOB Error' : forest.oob_score_}, ignore_index=True)
print('Now starting the normal random forest')
for i in range(1,300) :
    forest = RandomForestRegressor(n_estimators=i, warm_start=True, oob_score=True)
    forest.fit(X,y)

    # Calculate test error    
    error_sum = 0
    for index, row in X_test.iterrows() :
        y_pred = forest.predict(row.values.reshape(1,-1))
        error = abs(y_pred - y_test[index])
        error_sum += error
        
    size = len(X_test.index)
    average_error = error_sum/size
    forestDfInfo = forestDfInfo.append({'m' : 'n_features', 'B' : i, 'Test-Error' : average_error[0], 'OOB Error' : forest.oob_score_}, ignore_index=True)
    

forestDfInfo.head()
```
