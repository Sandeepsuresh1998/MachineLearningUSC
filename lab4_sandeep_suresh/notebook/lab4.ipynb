```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
df = pd.read_csv('../data/Youtube04-Eminem.csv');
df.head()
```

## Creating TF-IDF

```python
# Vectorizer for comments
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['CONTENT'])
X = X.toarray()

# Actually changing all comments to TDIDF score
for i, row in df.iterrows() :
    df.at[i,'CONTENT'] = sum(X[i])
df.tail()
```

```python
hamCount = len(df.loc[df.CLASS == 0])
spamCount = len(df.loc[df.CLASS == 1])

df1 = df.loc[df.CLASS == 0][:int(hamCount*0.8)]
df2 = df.loc[df.CLASS == 1][:int(spamCount*0.8)]
trainingDf = pd.concat([df1,df2], ignore_index=True)

df3 = df.loc[df.CLASS == 0][int(hamCount*0.8):]
df4 = df.loc[df.CLASS == 1][int(spamCount*0.8):]
testDf = pd.concat([df3,df4])
trainingDf.head()
```
